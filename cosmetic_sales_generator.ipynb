```python
import pandas as pd
import random
from datetime import datetime, timedelta

categories = ['Hair','Face','Eyes','Lips','Body','Skin Care']
products = {
    'Hair': ['Shampoo', 'Conditioner', 'Oil','Hair Mask','Hairstyling','Dry Shampoo'],
    'Face': ['Foundation','Concealer','Blush','Highlighter','Face Primer','Face Oil','Powder'],
    'Eyes': ['Eyeliner','Mascara','Eyeshadow','Eyelashes','Eye Primer'],
    'Lips': ['Lipstick','Lipgloss','Lip Liner','Lip Balm','Lip Stain'],
    'Body': ['Perfume','Soap','Body Wash','Sunscreen','Body Lotion'],
    'Skin Care': ['Facial Moisturizer', 'Facial Cleanser','Eye Treatment','Serum','Face Mask','Makeup Remover','Toner']
}

regions = ['North','South','East','West']
payments = ['Credit Card','Debit Card','Cash']

rows = []
start_date = datetime(2025, 1, 1)

for i in range(1000):
  category = random.choice(categories)
  product = random.choice(products[category])
  quantity = random.randint(1, 5)
  price = random.randint(10, 200)

  row = {
      'order_id': 1000 + i,
      'order_date': start_date + timedelta(days=random.randint(0,365)),
      'customer_id': f'C{random.randint(1,300):03}',
      'product_id': f'p{random.randint(100,200)}',
      'category': category,
      'product_name': product,
      'quantity': quantity,
      'price': price,
      'total_amount': quantity * price,
      'region': random.choice(regions),
      'payment_method': random.choice(payments)
  }

  rows.append(row)

df = pd.DataFrame(rows)

df.to_csv('cosmetics_sales_dataset.csv', index=False)

print('Dataset created: cosmetics_sales_dataset.csv')
```
